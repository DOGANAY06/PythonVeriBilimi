#### Ne yazık ki verimizde filmlerin IMDB puanları yok. Veri seti ile ayni satır uzunluğunda bir rastgele bir array yaratıp IMDB puanı olarak ekleyelim.

#### Puanları üretmek için alttaki random komutunu direk kullanabilirsiniz:

```python
IMDB = np.random.uniform(low=3.0, high=9.2, size=len(df['isim']))
IMDB = np.around(IMDB,decimals = 1)
```

```python
df["IMDB"]=IMDB
df
```

#### 'tip' ve 'kanal' sütunu arasinda nasil bir korelasyon vardir? (crosstab). En cok film ve dizi üretiminde katkisi olan kanal hangisidir?
- Bu korelasyonu bulmadan önce tip ve kanal sütunlarinin kac farkli degeri vardir. Onlari hesaplayalim.

```python
pd.crosstab(df["kanal"],df["tip"],margins=True).count()
```

```python
pd.crosstab(df["tip"],df["kanal"],margins=True).count()
```

#### 'tip' sütununda bulunan 'Movie' ve 'TV Show' degerlerini Türkce karsiliklari ile nasil degistirebiliriz?

```python

df.replace({"tip":{'Movie':'Film','TV Show':'Tv programi'}})
```

#### IMDB puan sütununu 'kanal' sütununa göre gruplandiralim (groupby) ve bu gruplandirma objesini 'grup' degiskeni olarak atayalim:

```python
ımdb_grup =df.groupby("IMDB")
ımdb_grup.agg(["describe"])["kanal"]
```

#### Kanallara göre gruplandirdigimiz bu grupta ortalama IMDB puanlarini hesaplayalim. 
- Kanallari ortalama IMDB puanlarina göre nasil siralayabiliriz? En yüksek ve en düsük IMDB ortalamalarina sahip kanallar hangisidir?

```python
kanal_grup =df.groupby("kanal")
kanal_grup.agg(["mean","max","min"])["IMDB"]
```

#### Oncelikle girilen degeri ikiye bölecek bir fonksiyon olusturalim. 
- Bu fonksiyonu kullanarak veri setimizdeki her film/dizinin IMDB puanini nasil yariya düsürebiliriz?

```python
def ikiyebol(arr):
    return arr/2.0
df["IMDB"].apply(ikiyebol)
```
